```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data_result = pd.read_csv('E:/Projek_python_kecil_kecil/deteksiObjekMuka/deteksiMuka/src/evaluasiModel/results.csv')
```

```python
fig, axs = plt.subplots(3,3, figsize=(15,15))

axs[0,0].plot(data_result['                  epoch'], data_result['         train/cls_loss'], label='train/cls_loss')
axs[0,0].plot(data_result['                  epoch'], data_result['           val/cls_loss'], label='val/cls_loss')
axs[0,0].legend()
axs[0,0].set_title("Class Loss")

axs[0,1].plot(data_result['                  epoch'], data_result['         train/dfl_loss'], label='train/dfl_loss')
axs[0,1].plot(data_result['                  epoch'], data_result['           val/dfl_loss'], label='val/dfl_loss')
axs[0,1].legend()
axs[0,1].set_title("DFL Loss")

axs[0,2].plot(data_result['                  epoch'], data_result['         train/box_loss'], label='train/box_loss')
axs[0,2].plot(data_result['                  epoch'], data_result['           val/box_loss'], label='val/box_loss')
axs[0,2].legend()
axs[0,2].set_title("BOX Loss")

axs[1,0].plot(data_result['                  epoch'], data_result['                 lr/pg0'], label='lr/pg0')
axs[1,0].legend()
axs[1,0].set_title("lr/pg0")

axs[1,1].plot(data_result['                  epoch'], data_result['                 lr/pg1'], label='lr/pg1')
axs[1,1].legend()
axs[1,1].set_title("lr/pg1")

axs[1,2].plot(data_result['                  epoch'], data_result['                 lr/pg2'], label='lr/pg2')
axs[1,2].legend()
axs[1,2].set_title("lr/pg2")

axs[2,0].plot(data_result['                  epoch'], data_result['      metrics/recall(B)'], label='metrics/recall(B)')
axs[2,0].legend()
axs[2,0].set_title("Recall")

axs[2,1].plot(data_result['                  epoch'], data_result['   metrics/precision(B)'], label='metrics/precision(B)')
axs[2,1].legend()
axs[2,1].set_title("Recall")

axs[2,2].plot(data_result['                  epoch'], data_result['       metrics/mAP50(B)'], label='metrics/mAP50(B)')
axs[2,2].legend()
axs[2,2].set_title("metrics/mAP50(B)")

plt.show()
```
